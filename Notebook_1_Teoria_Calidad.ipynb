## 1. Introducción: La Importancia de la Calidad
En la ciencia de datos, existe un concepto fundamental: **Garbage In, Garbage Out (GIGO)**. Si los datos de entrada son erróneos, incompletos o inconsistentes, cualquier análisis o modelo de Machine Learning producirá resultados poco fiables.



### ¿Qué define la 'Calidad'?
1. **Completitud:** ¿Faltan valores?
2. **Unicidad:** ¿Hay registros duplicados?
3. **Consistencia:** ¿Los formatos son iguales en todas las fuentes?
4. **Validez:** ¿Los datos siguen las reglas del negocio?

## 2. Los 8 Problemas Críticos de los Datos Reales
Basado en las diapositivas de clase, identificamos las situaciones que más afectan a los datasets empresariales:



1. **Valores Nulos:** Vacíos que rompen cálculos estadísticos.
2. **Filas Duplicadas:** Entradas repetidas que inflan KPIs.
3. **Tipos de Datos Mixtos:** Números guardados como texto.
4. **Cadenas sin Estandarizar:** Diferencias entre 'Cartagena' y 'cartagena'.
5. **Formatos de Fecha Inconsistentes:** Mezcla de formatos regionales.
6. **Problemas de Codificación:** Errores de lectura de caracteres (ñ, tildes).
7. **Delimitadores Incorrectos:** Confusión entre comas y puntos y comas.
8. **Campos Complejos:** Datos anidados que requieren limpieza avanzada.

## 3. Estrategias de Tratamiento

### A. Imputación de Datos
¿Qué hacemos cuando faltan datos? La teoría nos ofrece tres caminos:
- **Eliminar:** Si los nulos son pocos.
- **Imputar (Mediana):** Para variables numéricas (inmune a outliers).
- **Imputar (Moda):** Para variables categóricas.



### B. Detección de Outliers (Valores Atípicos)
Usamos el método del **Rango Intercuartílico (IQR)** para identificar datos que se salen de la lógica normal.
